```python
import pandas as pd
import numpy as np
```

```python
beike = pd.read_csv('./data.csv', encoding='utf-8-sig')
```

```python
beike['房屋户型']
```

```python
# 去除车位
beike_type=beike[beike['房屋户型']!='车位']
beike_type
```

```python
beike_type['房屋户型']
```

```python
beike_type['roomType1']=beike_type['房屋户型'].apply(lambda x: x.replace('室','室,').replace('厅','厅,').replace('厨','厨,'))
beike_type['roomType1']
```

```python
beike_type['roomType2']=beike_type['房屋户型'].apply(lambda x: x.replace('室','-').replace('厅','-').replace('厨','-').replace('卫',''))
beike_type['roomType2']
```

```python
beike_type['室']=beike_type['roomType2'].apply(lambda x: x.split('-')[0])
beike_type['厅']=beike_type['roomType2'].apply(lambda x: x.split('-')[1])
beike_type['厨']=beike_type['roomType2'].apply(lambda x: x.split('-')[2])
beike_type['卫']=beike_type['roomType2'].apply(lambda x: x.split('-')[3])
```

```python
beike_type
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(beike_type['roomType1'])
tfidf_matrix.shape
```

```python
print(tfidf_matrix)
```

```python
tfidf.vocabulary_
```

```python
tfidf_weight = tfidf_matrix.toarray()
tfidf_weight
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=100)
s = clf.fit(tfidf_weight)
```

```python
print(clf.inertia_)
```

```python
beike_type['labels']=clf.labels_
y=np.array(beike_type.loc[:,'labels'])
y
```

```python
beike_type
```

```python
x=np.array(beike_type.loc[:,('室','厅','厨','卫')])
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.4)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf1 = RandomForestClassifier(n_estimators=100)
clf1.fit(x_train,y_train)
```

```python
y_pred = clf1.predict(x_test)
```

```python
from sklearn import metrics
print('ACCURACY OF THE MODEL: ', metrics.accuracy_score(y_test, y_pred))
```

```python
a = int(input("室："))
b = int(input("厅："))
c = int(input("厨："))
d = int(input("卫：")) 
n=clf1.predict([[a,b,c,d]])
n
```

```python
return_house = beike_type[beike_type['labels']==int(n)].sort_values('成交时间',ascending=False).head(10).reset_index(drop=True)
return_house.iloc[:,1:20]
```
