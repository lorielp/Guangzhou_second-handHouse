```python
import pandas as pd
```

```python
beike = pd.read_csv('./data.csv', encoding='utf-8-sig')
```

```python
beike['房屋户型']
```

```python
# 去除车位
beike_type=beike[beike['房屋户型']!='车位']
beike_type
```

```python
beike_type['房屋户型']
```

```python
beike_type['type']=beike_type['房屋户型'].apply(lambda x: x.replace('室','室,').replace('厅','厅,').replace('厨','厨,'))
beike_type['type']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
# 构造tf-idf特征矩阵
tfidf_matrix = tfidf.fit_transform(beike_type['type'])
tfidf_matrix.shape
```

```python
print(tfidf_matrix)   # 3室-20 2厅-14 0厨-2 1卫-8
```

```python
tfidf.vocabulary_
```

```python
tfidf_weight = tfidf_matrix.toarray()
tfidf_weight
```
